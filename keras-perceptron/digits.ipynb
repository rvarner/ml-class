```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
from keras.datasets import mnist
import keras.models
import keras.layers
import keras.backend as K
import matplotlib.pyplot as plt
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape)
plt.imshow(X_train[0])
plt.show()
print(y_train.shape)
print(y_train[0])
```

```python
plt.rcParams['figure.figsize'] = (7,7)
offset = 0
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(X_train[i+offset], cmap='gray', interpolation='none')
    plt.title("label {}".format(y_train[i+offset]))
    plt.axis('off')
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(activation='relu',units=100))
model.add(keras.layers.Dense(activation='relu',units=50))
model.add(keras.layers.Dense(activation='softmax',units=10))
model.compile(loss='mse',optimizer='adam',metrics='accuracy')
```
